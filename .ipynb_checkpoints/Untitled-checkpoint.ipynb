```python
gross_mean=(data["Gross"].mean())
data['Gross']=data['Gross'].fillna(gross_mean)
```

```python
meta_score_mean=(data["Meta_score"].mean())
data['Meta_score']=data['Meta_score'].fillna(gross_mean)
```

```python
# Calculate the mode
sub_genre_mode = data['Subgenre'].mode()[0]

# Fill missing values in 'Subgenre' with the mode
data['Subgenre'] = data['Subgenre'].fillna(sub_genre_mode)
```

```python
# Calculate the mode
sub_genre_1_mode = data['Subgenre 1'].mode()[0]

# Fill missing values in 'Subgenre' with the mode
data['Subgenre 1'] = data['Subgenre 1'].fillna(sub_genre_1_mode)
```

```python
data['Certificate']=data['Certificate'].fillna('Unkown')
```

```python
data.info()
```

```python
data.to_csv("movies_data_clean.csv",index= False)
```
